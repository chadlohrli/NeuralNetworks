```python
#Libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Polynomial Regression

def l2Norm(x):
    l2 = np.sqrt(np.sum(np.power(x,2)))
    return l2

data = np.loadtxt('linear_regression_1_data.txt',dtype='float')
x = data[:,0].reshape(len(data),1)
y = data[:,1].reshape(len(data),1)

plt.plot(x,y)
plt.grid()

X = np.hstack((np.ones((len(x),1)), np.power(x,1), np.power(x,2)))
X_t = X.T

#Polynomial
sol = np.dot(np.linalg.inv(np.dot(X_t,X)),np.dot(X_t,y))
yhat = sol[0]+sol[1]*x+sol[2]*np.power(x,2)

plt.plot(x,y,label='data')
plt.plot(x,yhat,'b-',label='polynomial')
#plt.plot(x,yhat)
plt.title('Least square line fitting')
plt.xlabel('x')
plt.ylabel('y')

#caluclate L2 distance 
l2Distance = l2Norm(yhat-y)
print("L2 distance for polynomial regression: {}").format(l2Distance)

data = np.loadtxt('linear_regression_1_data.txt',dtype='float')
x = data[:,0].reshape(len(data),1)
y = data[:,1].reshape(len(data),1)

X = np.hstack((np.ones((len(x),1)),np.power(x,1)))
X_t = X.transpose((1,0))

#Linear
sol = np.dot(np.linalg.inv(np.dot(X_t,X)),np.dot(X_t,y))
yhat = sol[0]+sol[1]*x

plt.plot(x,yhat,'g-',label='linear')

#caluclate L2 distance 
l2Distance = l2Norm(yhat-y)
print("L2 distance for polynomial regression: {}").format(l2Distance)

plt.legend()
```

```python
# L1 vs L2 Loss Regression 

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data = np.loadtxt('linear_regression_2_data.txt',dtype='float')
x = data[:,0].reshape(len(data),1)
y = data[:,1].reshape(len(data),1)

X = np.hstack((np.ones((len(x),1)),np.power(x,1))) #(1000,2)

#print("shape of x:{}").format(x.shape)
#print("shape of X:{}").format(X.shape)

w = np.zeros((2,1))

plt.plot(x,y,'y-',label='Data')
plt.title('Least square line fitting')
plt.xlabel('x')
plt.ylabel('y')

rate = 0.00001
iterations = 2000


#gradient descent for L1 Loss Regression
for i in range(0,iterations):
    a = np.sum(np.sign(np.dot(X,w) - y) * X,axis=0).reshape(2,1)
    w = w - rate*a
    
print("L1 weight vector: {}").format(w)
yhat = w[0] + w[1]*x;
plt.plot(x,yhat,'r-',label='L1')

#L2 Loss Regression
X_t = X.transpose((1,0))
sol = np.dot(np.linalg.inv(np.dot(X_t,X)),np.dot(X_t,y))

print("L2 weight vector: {}").format(sol)
yhat = sol[0]+sol[1]*x
plt.plot(x,yhat,'b-',label='L2')


plt.legend()
plt.show()
```
