```python
#Libraries
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
#Problem 3

import numpy as np

#data processing
data = np.loadtxt('logistic_regression_data.txt', delimiter=',', usecols=[0,1,2,3])
d1 = np.insert(data[0:50,:], 4, 0, axis=1) #Iris-versicolor = 0
d2 = np.insert(data[50:100,:], 4, 1, axis=1) #Iris-verginica = 1

test = np.vstack((d1[0:15,:],d2[0:15,:]))
train = np.vstack((d1[15:50,:],d2[15:50,:]))

xtrain = train[:,0:4]
ytrain = train[:,4:5]

xtest = test[:,0:4]
ytest = test[:,4:5]

# --PCA Data Plotting-- #

target_names = ['Iris-versicolor', 'Iris-verginica']

pca = PCA(n_components=2)
Xplot = pca.fit(xtest).transform(xtest)
yplot = ytest.reshape(30,)

plt.figure(0)
colors = ['navy', 'turquoise']
lw = 2

for color, i, target_name in zip(colors, [0, 1], target_names):
    plt.scatter(Xplot[yplot == i, 0], Xplot[yplot == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of IRIS dataset')


def activation(z):
    g  = 1/(1 + np.exp(-1*z))
    return g 
            
def forward(x,w,b):
    z = np.dot(x,w) + b
    return activation(z)

def costf(y,h):
    a = np.log(h).reshape(70,1)
    b = np.log(1-h).reshape(70,1)
    cost = -1*(np.sum(y*a + (1-y)*b))
    return cost


def gradientDescent(x, y, w, b, alpha, m, iters):
    serr = []
    print("Training...")
    for i in range(0, iters):
        h = forward(x,w,b)
        err = h - y.reshape(70,)
        cost = costf(y,h) 
        dw = np.dot(x.T, err)  
        w = w - alpha*dw 
        b = b - alpha*(sum(err))
        if i % 1000 == 0:
            print("error:{}").format(sum(err)/m)
        serr.append(cost)
    return w, b, serr 

m,n = xtrain.shape
w = np.ones(n)
b = 1
alpha = 0.001
iters = 8000

w, b, cost = gradientDescent(xtrain,ytrain,w,b,alpha,m,iters)
#print serr
plt.figure(1)
plt.plot(cost,label='Cost Function')
plt.xlabel('iterations')
plt.ylabel('cost')
plt.legend()

print("")
print "w:",w
print "b:",b
print "cost:", cost[iters-1]

# Testing 

def predict(h):
    if(h >= 0.5):
        return 1
    else:
        return 0

#Define Metric Parameters
tp = 0.
fn = 0.
tn = 0.
fp = 0.

m,n = xtest.shape
serr = []

for i in range(0,m):
    h = forward(xtest[i],w,b)
    p = predict(h)
    err = p - ytest[i]
    serr.append(err/m)
    #Error Metrics
    if(ytest[i] == 1 and p == 1):
        tp += 1
    elif(ytest[i] == 1 and p == 0):
        fn += 1
    elif(ytest[i] == 0 and p == 0):
        tn += 1
    elif(ytest[i] == 0 and p == 1):
        fp += 1  
    
plt.figure(2)
plt.plot(serr,label="Test Error")
plt.xlabel('iterations')
plt.ylabel('error')
plt.legend()

    
precision = tp / (tp + fp)
recall = tp / (tp + fn)
fscore = 2 * (recall * precision)/(recall + precision)
accuracy = (tp + tn) / (tp + tn + fn + fp)

print("")
print("--Decision Boundary--")
print("{}*x1 + {}*x2 + {}*x3 + {}*x4 = {}").format(w[0],w[1],w[2],w[3],-b)
print("")
print("--Test Error Metrics--")
print("precision:{}, recall:{:0.2f}, fscore:{:0.2f}, accuracy:{}").format(precision,recall,fscore,accuracy)
    
  
```
